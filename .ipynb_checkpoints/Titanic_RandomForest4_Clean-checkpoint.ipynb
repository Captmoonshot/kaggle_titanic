# Would You Have Survived on The Titanic?

One of the most famous/infamous examples of Black Swans (rare events that have outsized impacts) in history is the sinking of the Titanic on April 15th 1912 after she collided with an iceberg resulting in the death of 1502 men, women, and children out of 2224 passengers.

According to the [kaggle website](https://www.kaggle.com/c/titanic), "Although there was some luck involved in surviving the sinking, some groups of people were more likey to survive than others, such as women, children, and the upper-class."

For this challege we want to find out who were most likely to survive or die on the Titanic based on some of the characteristics of the passengers.

You can get the datasets for training and testing from the link mentioned above.

Let's get to it.

The ultimate goal is to try get an accurate set predictions in a csv file format with 2 columns:
* PassengerId (sorted in any order)
* Survived (consisting of 0s and 1s, where 1=survived and 0=deceased)

### Data Dictionary

| Variable      | Definition           | Key                                      |
| ------------- | --------------------:| ----------------------------------------:|
| survival      | Survival             | 0 = No, 1 = Yes                          |
| pclass        | Ticket class         | 1 = 1st, 2 = 2nd, 3 = 3rd                |
| sex           | Sex                  |                                          |
| Age           | Age in years         |                                          |
| sibsp         | # of siblings/spouses|                                          |
| parch         | # of parents/children|                                          |
| ticket        | Ticket numbe         |                                          |
| fare          | Passenger fare       |                                          |
| cabin         | Cabin number         |                                          |
| embarked      | port of embarkation  | C=Cherbourg, Q=Queenstown, S=Southampton |

## Our Baseline Model:

At this point, for purposes of brevity, I'm going to go immediately into fitting and evaluating a baseline model with which to do compare our other models.  

For the baseline model I'm going to run a Random Forest which is nice because we don't have to do special preprocessing of the data like scaling and one-hot-encoding.

```python
# Pop off the dependent variable
y = data.pop('Survived')
```

```python
# Split the training dataset into separate training and test sets 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
```

```python
print(list(X_train.columns))
```

```python
print(list(X_test.columns))
```

We want to run the baseline model on just the numerical variables

```python
X_train_numeric_variables = list(X_train.dtypes[X_train.dtypes != "object"].index)
X_train[X_train_numeric_variables].head()
```

```python
X_test_numeric_variables = list(X_test.dtypes[X_test.dtypes != "object"].index)
X_test[X_test_numeric_variables].head()
```

```python
# Running RandomForestClassifier on just the numerical variables of the training and test set

from sklearn.ensemble import RandomForestClassifier

baseline_model = RandomForestClassifier(n_estimators=100, random_state=42)

baseline_model.fit(X_train[X_train_numeric_variables], y_train)

print("Training accuracy score: {:.2f}".format(baseline_model.score(X_train[X_train_numeric_variables], y_train)))
print("Testing accuracy score: {:.2f}".format(baseline_model.score(X_test[X_test_numeric_variables], y_test)))
```

This score for a baseline model using only the numerical variables isn't too bad.

But now we're going to work on the categorical variables and rerun the Random Forest model to see if we can get a higher accuracy score.

```python
def clean_cabin(X_train):
    try:
        return X_train[0]
    except TypeError:
        return "None"
X_train["Cabin"] = X_train.Cabin.apply(clean_cabin)
```

```python
def clean_cabin(X_test):
    try:
        return X_test[0]
    except TypeError:
        return "None"
X_test["Cabin"] = X_test.Cabin.apply(clean_cabin)
```

I'm also going to drop two categorical variables that probably isn't too helpful for predictive modeling

```python
X_train.drop(["Name", "Ticket"], axis=1, inplace=True)
X_train.head()
```

```python
X_test.drop(["Name", "Ticket"], axis=1, inplace=True)
X_test.head()
```

Here we one-hot-encode our categorical variables

```python
categorical_variables = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    X_train[variable].fillna("Missing", inplace=True)
    # Create array of dummies
    dummies = pd.get_dummies(X_train[variable], prefix=variable)
    # Update X to inlude dummies and drop the old variable
    X_train = pd.concat([X_train, dummies], axis=1)
    X_train.drop([variable], axis=1, inplace=True)
```

We create the same function for X_test

```python
categorical_variables = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    X_test[variable].fillna("Missing", inplace=True)
    # Create array of dummies
    dummies = pd.get_dummies(X_test[variable], prefix=variable)
    # Update X to inlude dummies and drop the old variable
    X_test = pd.concat([X_test, dummies], axis=1)
    X_test.drop([variable], axis=1, inplace=True)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
```

We see from the shapes that X_test lacks some of the columns of X_train.  I'm just going to drop the extra columns from the X_train to make it simple.

```python
X_train.drop(["Cabin_A", "Cabin_T", "Embarked_Missing"], axis=1, inplace=True)
X_train.head()
```

```python
print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
```

Now we run the RandomForestClassifier again and evaluate the model

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)

print("X Training accuracy score: {:.2f}".format(model.score(X_train, y_train)))
print("X Testing accuracy: {:.2f}".format(model.score(X_test, y_test)))
```

```python
# Running the model with some different parameters
model2 = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=42, max_features="auto", min_samples_leaf=5)
model2.fit(X_train, y_train)

print("X Training accuracy score: {:.2f}".format(model2.score(X_train, y_train)))
print("X Testing accuracy: {:.2f}".format(model2.score(X_test, y_test)))
```

This is definitely an improvement over the previous scores.  But now I want to scale the data so we can run some other models to compare with on the dataset.

```python
# We're scaling on just the numerical variables
# Remember we fit_transform on training set and transform on testing set

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']] = scaler.fit_transform(X_train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
X_test[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']] = scaler.transform(X_test[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
```

```python
X_train.head()
```

```python
X_test.head()
```

### Support Vector Machine

```python
from sklearn.svm import SVC

svm_clf = SVC(C=1000, random_state=42)
svm_clf.fit(X_train, y_train)
```

```python
print("SVM training accuracy score: {:.2f}".format(svm_clf.score(X_train, y_train)))
print("SVM testing accuracy score: {:.2f}".format(svm_clf.score(X_test, y_test)))
```

### Multilayer Perceptron

```python
from sklearn.neural_network import MLPClassifier

mlp_clf = MLPClassifier(max_iter=1000, alpha=0.001, random_state=42)
mlp_clf.fit(X_train, y_train)
```

```python
print("MLP Training accuracy score: {:.2f}".format(mlp_clf.score(X_train, y_train)))
print("MLP Testing accuracy score: {:.2f}".format(mlp_clf.score(X_test, y_test)))
```

### K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)
```

```python
print("KNN training accuracy score: {:.2f}".format(knn_clf.score(X_train, y_train)))
print("KNN testing accuracy score: {:.2f}".format(knn_clf.score(X_test, y_test)))
```

## Final Model

It seems as if the neural network model (multilayer perceptron) is giving us the best accuracy rates over the others so that's what we're going to go with as the final model and this time we will be preprocessing the whole dataset (data, y).

We already took care of missing numerical values by imputing with mean so we don't need to worry about that here.  We will need to do all the other things we did with X_train, X_test like dropping misc. variables, scaling, and one-hot-encoding categorical variables

```python
data.head()
```

```python
data_test.head()
```

```python
data_test.head()
```

```python
data.head()
```

```python
# Scaling numerical variables
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']] = scaler.fit_transform(data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
data.head()
```

```python
data_test[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']] = scaler.transform(data_test[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
```

```python
data_test.head()
```

```python
# work on categorical variables

def clean_cabin(data):
    try:
        return data[0]
    except TypeError:
        return "None"
data["Cabin"] = data.Cabin.apply(clean_cabin)

data.head()
```

```python
def clean_cabin(data_test):
    try:
        return data_test[0]
    except TypeError:
        return "None"
data_test["Cabin"] = data_test.Cabin.apply(clean_cabin)

data_test.head()
```

```python
categorical_variables = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_variables:
    # Fill in missing data with the word "Missing"
    data[variable].fillna("Missing", inplace=True)
    dummies = pd.get_dummies(data[variable], prefix=variable)
    # Update data to include dummies and drop the old variable
    data = pd.concat([data, dummies], axis=1)
    data.drop([variable], axis=1, inplace=True)
    
data.head()
```

```python
categorical_variables = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_variables:
    # Fill missing data with the word "Missing"
    data_test[variable].fillna("Missing", inplace=True)
    # Create array of dummies
    dummies = pd.get_dummies(data_test[variable], prefix=variable)
    # Update data_test to include dummies and drop the old variable
    data_test = pd.concat([data_test, dummies], axis=1)
    data_test.drop([variable], axis=1, inplace=True)
    
data_test.head()
```

```python
print("data shape: {}".format(data.shape))
print("data_test shape: {}".format(data_test.shape))
```

It looks like the data_test set is missing some columns so we're going to drop those from data training set

```python
data.drop(["Cabin_T", "Embarked_Missing"], axis=1, inplace=True)

print("data shape: {}".format(data.shape))
print("data_test shape: {}".format(data_test.shape))

data.head()
```

```python
# And now for the final MLP model:

from sklearn.neural_network import MLPClassifier

mlp_clf_final = MLPClassifier(max_iter=1000, alpha=0.001, random_state=42)
mlp_clf_final.fit(data, y)
```

```python
print("Multilayer Perceptron Training accuracy score: {:.2f}".format(mlp_clf_final.score(data, y)))
```

```python
# Predictions
mlp_clf_final_predictions = mlp_clf_final.predict(data_test)
mlp_clf_final_predictions
```

```python
# Ready for Kaggle submission
my_titanic_submission = pd.DataFrame({'PassengerId': data_test.index, 'Survived': mlp_clf_final_predictions})

my_titanic_submission.to_csv('Titanic_submission13_mlp.csv', index=False)
```
